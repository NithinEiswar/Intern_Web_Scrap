```python
# You tell Python that you want to use a library with the import statement.
import requests
```

```python
# Get the https://www.tnagrisnet.tn.gov.in/Contact.php page
req = requests.get("http://nrcb.res.in/list-of-progressive-farmers.php")
```

```python
dir(req)
```

```python
page = req.text
page
```

```python
from bs4 import BeautifulSoup
```

```python
# Your Code Here
soup = BeautifulSoup(page, 'html.parser')
```

```python
soup
```

```python
title = soup.find('title')
print(title) # Prints the tag
print(title.string) # Prints the tag string content
```

```python
soup.table
```

```python
table_req = soup.find_all("table")
```

```python
print(table_req)
```

```python
rows=table_req[0].tbody.find_all("tr")
print(rows)
```

```python
columns = []
for th in table_req[0].find_all("th"):
        # remove any newlines and extra spaces from left and right
        columns.append(th.text)    
print(columns)  
```

```python
import pandas as pd
```

```python
values= []
i=0
while i<237:
    for td in rows[i].find_all("td"):
        try:
            values.append(td.text)
        except:
            values.append(None)
    i=i+1       
       
print(values) 
```

```python
stacked_values = zip(*[values[i::4] for i in range(len(columns))])
stacked=list(stacked_values)
#res = [list(ele) for ele in list(stacked_values)] 
#print (str(res))
print(stacked)
```

```python
df=pd.DataFrame(list(stacked),columns=columns)
df['S.No'] = [i[:-1] for i in df['S.No']] 
df['S.No'] = [int(i) for i in df['S.No']] 
df['Name & Address'] = [s.strip() for s in df['Name & Address']] 
df['Name & Address'] = [s.split(",", 1)[0] for s in df['Name & Address']] 
df['Name & Address'] = [s.split("\r", 1)[0] for s in df['Name & Address']] 
df['Phone Number'] = [s.strip() for s in df['Phone Number']] 
df['District'] = [s.strip() for s in df['District']] 
df.set_index('S.No',inplace=True)
df.tail(14)
```

```python
df.to_csv('NCRB.csv')
```
