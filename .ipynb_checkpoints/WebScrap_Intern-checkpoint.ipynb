```python
import camelot
import pandas as pd
data=pd.DataFrame()
frames=[]
tables = camelot.read_pdf("Tamil-Nadu_2015 (1).pdf",pages='all')
for i in list(range(tables.n)):
    tables[i].df = tables[i].df.replace('\n','', regex=True)
    data=data.append(tables[i].df)
tables.export('foo.csv',f='csv')
```
